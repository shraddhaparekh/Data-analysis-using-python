# Transformation Functions

```python
#Map and Apply
#Changes/modify certain values
#Map is used for small changes(one line code changes)
#Apply is used for larger (>1 line code changes)
```

```python
add5 = lambda x:x+5
add5(30)
```

```python
sub2 = lambda x:x-2
sub2(15)
```

```python
video_games.head()
```

```python
video_games.NA_Sales =video_games.NA_Sales.map(lambda x:x+5)
video_games.head()
```

```python
video_games.EU_Sales =video_games.EU_Sales.map(lambda x:x+5)
video_games.head()
```

```python
#NA_Sales and EU_Sales both must be reduced by 5
def sub5(row):
  row.NA_Sales -=5
  row.EU_Sales -=5
  return row

video_games = video_games.apply(sub5, axis = 1)
```

```python
video_games.head()
```

# Grouping & Sorting

```python
# I wanna know which Genre has many games
# Which genre has most games?
video_games.groupby("Genre").Genre.count()
```

```python
#which developer has most games sold
dev_games_count = video_games.groupby("Developer").Name.count()
```

```python
dev_games_count
```

```python
dev_games_count.sort_values(ascending = False)
```

```python
#which developer has made how much sales in USD globally?
video_games.groupby("Developer").Global_Sales.sum().sort_values(ascending = False)
```

```python
#sort games based on their Genre
video_games.sort_values("Genre",ascending = False)
```

```python
#Sort Games by genre and under each Genre , we need to have games in alphabetical order
video_games.sort_values(["Genre","Name"],ascending =[False,True])
```
